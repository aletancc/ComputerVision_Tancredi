```python
import matplotlib.pyplot as plt
import os
from PIL import Image

# Path to the correct output folder where YOLOv5 saved the images
output_folder = '/content/runs/detect/exp2'

# List all the images in the output folder
output_images = [img for img in os.listdir(output_folder) if img.endswith('.jpg')]

# Display the first few images with bounding boxes
for i, image_name in enumerate(output_images[:5]):  # Change the range if you want more images
    img_path = os.path.join(output_folder, image_name)
    img = Image.open(img_path)
    plt.figure(figsize=(10, 10))
    plt.imshow(img)
    plt.axis('off')  # Hide axis
    plt.show()

```

```python
import os
from PIL import Image

# Path to the correct output folder
output_folder = '/content/yolov5/yolov5/runs/detect/exp3'

# List all files in the output folder
output_files = os.listdir(output_folder)

# Check the file types and validity of each image
for file_name in output_files:
    file_path = os.path.join(output_folder, file_name)
    try:
        img = Image.open(file_path)
        img.verify()  # Verify that it's an image file
        print(f"{file_name} is a valid image.")
    except (IOError, SyntaxError) as e:
        print(f"{file_name} is not a valid image file: {e}")
```
